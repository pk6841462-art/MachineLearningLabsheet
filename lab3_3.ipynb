```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn import tree
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv(r"D:\prashant\lab3_3.csv")
df
```

```python
input=df.drop(['pass'],axis='columns')
target = df['pass']
```

```python
le_attendence=LabelEncoder()
le_studyhour=LabelEncoder()
le_sleep=LabelEncoder()
le_internetuse=LabelEncoder()
```

```python
input['attendence_n']=le_attendence.fit_transform(input['attendence'])
input['studyhour_n']=le_studyhour.fit_transform(input['studyhour'])
input['sleep_n']=le_sleep.fit_transform(input['sleep'])
input['internetuse_n']=le_internetuse.fit_transform(input['internetuse'])
```

```python
input_n=input.drop(['attendence','studyhour','sleep','internetuse'],axis='columns')
input_n
```

```python
model=tree.DecisionTreeClassifier()
model.fit(input_n,target)
```

```python
prediction=model.predict([[0,0,0,0]])
print("prediction:",prediction)
prediction=model.predict([[1,1,1,1]])
print("prediction:",prediction)
prediction=model.predict([[2,2,2,2]])
print("prediction:",prediction)

```

```python
accuracy=model.score(input_n,target)
print("model accuracy:",accuracy)
```

```python
plt.figure(figsize=(8,10))
tree.plot_tree(model,feature_names=input_n.columns,class_names=["NO","YES"],filled=True)
plt.show()
```
