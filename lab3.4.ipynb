```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn import tree
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv(r"D:\prashant\lab3_4.csv")
df
```

```python
input=df.drop(['flu'],axis='columns')
target = df['flu']
```

```python
le_fever=LabelEncoder()
le_cough=LabelEncoder()
le_headeche=LabelEncoder()
le_bodypain=LabelEncoder()
```

```python
input['fever_n']=le_fever.fit_transform(input['fever'])
input['cough_n']=le_cough.fit_transform(input['cough'])
input['headeche_n']=le_headeche.fit_transform(input['headeche'])
input['bodypain_n']=le_bodypain.fit_transform(input['bodypain'])
```

```python
input_n=input.drop(['fever','cough','headeche','bodypain'],axis='columns')
input_n
```

```python
model=tree.DecisionTreeClassifier()
model.fit(input_n,target)
```

```python
prediction=model.predict([[0,0,0,1]])
print("prediction:",prediction)
```

```python
accuracy=model.score(input_n,target)
print("model accuracy:",accuracy)
```

```python
plt.figure(figsize=(8,10))
tree.plot_tree(model,feature_names=input_n.columns,class_names=["NO","YES"],filled=True)
plt.show()
```
