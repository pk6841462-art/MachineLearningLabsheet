```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier
```

```python
df=pd.read_csv(r"D:\prashant\lab4_2.csv")
df
```

```python
plt.xlabel('GlucoseLevel')
plt.ylabel('BMI')
plt.scatter(df.GlucoseLevel,df.BMI,c=df.Diabetic)
plt.show()
```

```python
model=KNeighborsClassifier(n_neighbors=1)
model.fit(df[['GlucoseLevel','BMI']],df.Diabetic)
```

```python
prediction=model.predict([[120,25]])
print("prediction:",prediction)
```
