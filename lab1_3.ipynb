```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df=pd.read_csv(r"D:\prashant\study hours predict.csv")
df
```

```python
%matplotlib inline
plt.xlabel('study')
plt.ylabel('marks')
plt.scatter(df.study, df.marks, color='black', marker='*')
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['study']],df.marks)
```

```python
reg.predict([[8]])
```

```python
print("slope:",reg.coef_)
print("slope:",reg.intercept_)
```
