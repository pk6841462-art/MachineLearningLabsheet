```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn import tree
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv(r"D:\prashant\job.csv")
df
```

```python
input =df.drop(['salary_100k'],axis='columns')
target=df.salary_100k
```

```python
le_company=LabelEncoder()
le_job=LabelEncoder()
le_degree=LabelEncoder()
```

```python
input['company_n']=le_company.fit_transform(input['company'])
input['job_n']=le_job.fit_transform(input['job'])
input['degree_n']=le_degree.fit_transform(input['degree'])
```

```python
input_n=input.drop(['company','job','degree',],axis='columns')
```

```python
model=tree.DecisionTreeClassifier()
model.fit(input_n,target)
```

```python
prediction=model.predict([[2,2,2]])
print("prediction:",prediction)
```

```python
accuracy=model.score(input_n,target)
print("model accuracy:",accuracy)
```

```python
plt.figure(figsize=(8,10))
tree.plot_tree(model,feature_names=input_n.columns,class_names=["NO","YES"],filled=True)
plt.show()
```
